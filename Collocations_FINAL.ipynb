```python
import pandas as pd

coll = pd.read_csv("collocations_4.csv", header=0)
coll.head()
```

```python
coll.dtypes
```

```python
import nltk
from nltk.collocations import *

bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
from nltk.metrics import ContingencyMeasures
cont_bigram_measures = ContingencyMeasures(bigram_measures)
```

```python
arr = []
for index, row in coll.iterrows():
   ch = cont_bigram_measures.pmi(row["n_ii"], row["n_io"], row["n_oi"], row["n_oo"])
   arr.append('%0.2f'% ch)
print(arr)
```

```python
s1 = pd.Series(arr, name='PMI')
result = pd.concat([coll, s1], axis=1)
result
```

```python
result.to_csv('result1.csv')
```

```python
chi = []
for index, row in coll.iterrows():
   sq = cont_bigram_measures.chi_sq(row["n_ii"], row["n_io"], row["n_oi"], row["n_oo"])
   chi.append('%0.2f'% sq)
print(chi)
s2 = pd.Series(chi, name='X^2')
result2 = pd.concat([result, s2], axis=1)
result2
```

```python
stud_arr = []
for index, row in coll.iterrows():
   stud_t = cont_bigram_measures.student_t(row["n_ii"], row["n_io"], row["n_oi"], row["n_oo"])
   stud_arr.append('%0.2f'% stud_t)
#print(stud_arr)
s3 = pd.Series(stud_arr, name='student_t')
result3 = pd.concat([result2, s3], axis=1)
result3
```

```python
LL_arr = []
for index, row in coll.iterrows():
   LL = cont_bigram_measures.likelihood_ratio(row["n_ii"], row["n_io"], row["n_oi"], row["n_oo"])
   LL_arr.append('%0.2f'% LL)
#print(stud_arr)
s4 = pd.Series(LL_arr, name='Likelihood')
final= pd.concat([result3, s4], axis=1)
final
```

```python
writer = pd.ExcelWriter('pandas_simple.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
final.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
